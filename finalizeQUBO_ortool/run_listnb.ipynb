```python
import papermill as pm
import ipykernel
# Danh sách các file cần chạy
notebooks = [
    'adaptive_box_ortool_raw_simple.ipynb',
    'adaptive_box_ortool_raw_simple2.ipynb',
    'adaptive_box_ortool_raw_simple3.ipynb',
    'adaptive_box_ortool_raw_simple_vs10.ipynb',
    'adaptive_box_ortool_raw_simple2_vs10.ipynb',
    'adaptive_box_ortool_raw_simple3_vs10.ipynb',
    'adaptive_box_ortool_raw_simple_2v.ipynb',
    'adaptive_box_ortool_raw_simple2_2v.ipynb',
    'adaptive_box_ortool_raw_simple3_2v.ipynb',
]
current_kernel = ipykernel.connect.get_connection_file().split("-")[-1].split(".")[0]
for nb in notebooks:
    print(f">>> Đang chạy {nb} bằng kernel: {current_kernel}")
    try:
        pm.execute_notebook(
            input_path=nb,
            output_path=nb,
            kernel_name='python3' # Sử dụng kernel của Anaconda
        )
        print(f"OK: {nb} đã lưu.\n")
    except Exception as e:
        print(f"Lỗi ở file {nb}: {e}")
```
